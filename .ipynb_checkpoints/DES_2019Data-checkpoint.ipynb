```python
import pandas as pd
```

```python
dSim = pd.read_json("Data\\vesseldata\\2019Sim.json")
```

```python
dSim['LOCALTIME'] = pd.to_datetime(dSim['LOCALTIME'], format = '%Y-%m-%dT%H:%M:%S')
dSim['UTCTIME'] = pd.to_datetime(dSim['UTCTIME'], format = '%Y-%m-%dT%H:%M:%S')
```

```python
dSim
```

```python
a = dSim.drop([0,2,3])
a = a.reset_index(drop = True)
```

```python
a
```

### Get Interval

```python
b = a.loc[a['MOVE'] == 0].reset_index(drop = True)
```

```python
interval = list()
for i in list(range(1, len(b['UTCTIME']))) : 
    interval.append(b['UTCTIME'][i]-b['UTCTIME'][i-1])
```

```python
len(interval)
```

```python
intHours = list()
for i in interval:
    intHours.append((i.total_seconds()/3600))
```

```python
interval.append('n/a')
intHours.append('n/a')
```

### Get Service Time

```python
results = pd.DataFrame()
for i in (a['NAME'].unique()):
    df = a.loc[a['NAME'] == i]
    dfin = (df.loc[df['MOVE'] == 0]).reset_index(drop = True)
    dfout= (df.loc[df['MOVE'] == 1]).reset_index(drop = True)
    dfin['LEAVING'] = dfout['UTCTIME']
    dfin['SERVICE'] = dfout['UTCTIME'] - dfin['UTCTIME']
    df = dfin.append(dfout)
    results = pd.concat([results, df], axis=0).reset_index(drop=True)
```

```python
results.head(60)
```

```python
c = results[results['MOVE'] == 0]
c = c.sort_values(by=['UTCTIME'])
```

```python
c['INTERVAL'] = interval
c = c.reset_index(drop = True)
```

```python
c
```

### Convert to hours

```python
serHours = list()
for i in c['SERVICE']:
    serHours.append((i.total_seconds()/3600))
```

```python
c['SERVICEHOURS'] = serHours
c['INTERVALHOURS'] = intHours
```

```python
c.head(60)
```

```python
c.tail(60)
```

```python
os.getcwd()
```

```python
c.to_json('2019_progress.json')
```
