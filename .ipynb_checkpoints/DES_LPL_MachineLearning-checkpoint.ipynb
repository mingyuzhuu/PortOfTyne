```python
dTime = pd.read_csv('TimeDelta.csv')
```

```python
dTime
```

```python
dTime['ARRIVING'] = pd.to_datetime(dTime['ARRIVING'], format = '%Y-%m-%dT%H:%M:%S')
dTime['LEAVING'] = pd.to_datetime(dTime['LEAVING'], format = '%Y-%m-%dT%H:%M:%S')
```

```python
dTime
```

```python
type(dTime['LEAVING']-dTime['ARRIVING'])
```

```python
dTime['SERVICE'] = dTime['LEAVING']-dTime['ARRIVING']
```

```python
dTime
```

```python
arriving = dTime['ARRIVING']
```

```python
arriving
```

```python
interval = list()
for i in list(range(1, len(arriving))) : 
    interval.append(arriving[i]-arriving[i-1])
```

```python
interval.append(interval[30])
```

```python
interval
```

```python
dTime['INTERVAL'] = interval
```

```python
dTime
```

```python
dTime.to_json('LPL_Vessel_2019.json')
```

```python
serHours = list()
for i in dTime['SERVICE']:
    serHours.append((i.total_seconds()/3600))
```

```python
intHours = list()
for i in dTime['INTERVAL']:
    intHours.append((i.total_seconds()/3600))
```

```python
dTime['SERVICEHOURS'] = serHours
```

```python
dTime['INTERVALHOURS'] = intHours
```

```python
dTime
```

```python
plt.hist(intHours,bins=12, alpha=1)
```

```python
dTime['HandlingRate'] = dTime['SHIPLOAD'] / dTime['SERVICEHOURS']
```

```python
dTime
```

```python
dTime.to_json('LPL_Vessel_2019.csv')
```

```python
dTime['Grab'] = dTime['HandlingRate'] / 72
```

```python
dTime
```

```python
dRate = dTime.drop(columns=['ARRIVING','LEAVING','SERVICE','INTERVAL','SERVICEHOURS','INTERVALHOURS','Grab'])
```

```python
dRate
```

```python
dRate = dRate[dRate.SHIPLOAD != 0.00000]
```

```python
dRate
```

```python
dRate['GT'] = np.log(dRate['GT'])
dRate['DWT'] = np.log(dRate['DWT'])
dRate['SHIPLOAD'] = np.log(dRate['SHIPLOAD'])
dRate['HandlingRate'] = np.log(dRate['HandlingRate'])
```

```python
dRate
```

```python
dReg = dRate
dReg
```

```python
dReg.plot.scatter(x='SHIPLOAD',y='HandlingRate')
```

```python
from sklearn import linear_model
```

```python
X = dTime[['GT','DWT','SHIPLOAD']]
y = dTime['HandlingRate']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
print(regr.coef_)
```

```python
regr.intercept_
```

```python
y_pre = regr.predict(X)
```
