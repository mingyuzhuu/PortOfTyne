```python
import pandas as pd
import json
import urllib as u
import datetime
```

```python
#create APL link 
domain='http://api.worldweatheronline.com/premium/v1/past-weather.ashx'
```

### ==== 2018

```python
value2018 = {'q':'Jarrow','date':'2020-01-01','enddate':'2020-01-31','tp':'24','format':'json','key':'2428c01539614ee9bd6142524221802'}
prams2018 = u.parse.urlencode(value2018)
prams2018
```

```python
url2018 = domain +'?' + prams2018
url2018
```

```python
df = pd.DataFrame(json.loads(u.request.urlopen(url2018).read()))
```

```python
df = pd.DataFrame(df['data']['weather'])
df
```
