### Get Interval

```python
b = a.loc[a['MOVE'] == 0].reset_index(drop = True)
```

```python
interval = list()
for i in list(range(1, len(b['UTCTIME']))) : 
    interval.append(b['UTCTIME'][i]-b['UTCTIME'][i-1])
```

```python
len(interval)
```

```python
intHours = list()
for i in interval:
    intHours.append((i.total_seconds()/3600))
```

### Get Service

```python
results = pd.DataFrame()
for i in (a['NAME'].unique()):
    df = a.loc[a['NAME'] == i]
    dfin = (df.loc[df['MOVE'] == 0]).reset_index(drop = True)
    dfout= (df.loc[df['MOVE'] == 1]).reset_index(drop = True)
    dfin['LEAVING'] = dfout['UTCTIME']
    dfin['SERVICE'] = dfout['UTCTIME'] - dfin['UTCTIME']
    df = dfin.append(dfout)
    results = pd.concat([results, df], axis=0).reset_index(drop=True)
```
